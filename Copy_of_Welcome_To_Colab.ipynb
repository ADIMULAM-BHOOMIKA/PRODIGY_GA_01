```
from transformers import pipeline

generator = pipeline("text-generation", model="gpt2")

result = generator(
    "Artificial Intelligence is",
    max_length=50,
    num_return_sequences=1
)

print(result[0]['generated_text'])
```

```
from transformers import pipeline

generator = pipeline("text-generation", model="gpt2")

result = generator(
    "I am a B.Tech student",
    max_length=30,
    num_return_sequences=1
)

print(result[0]["generated_text"])
```

```
with open("data.txt", "w") as f:
    f.write("""
Bhoomika is a B.Tech student.
She likes Python programming.
She is learning Generative AI.
She builds projects and uploads them to GitHub.
""")

print("File created successfully!")
```

```
with open("data.txt", "r") as f:
    print(f.read())
```

```
with open("data.txt", "r") as f:
    custom_text = f.read()

prompt = custom_text + "\nBhoomika"

result = generator(
    prompt,
    max_new_tokens=50,
    num_return_sequences=1
)

print(result[0]["generated_text"])
```
